###Encapsulation:

Individual objects have their own, seperate internal state.
It's therefore important to distinguish between:  
1. **classes**:  Templates from which objects are created
2. **objects** or **instances**:  (synonymous) Objects created from classes

**A few examples of classes:  **

    list
    dict
    tuple
    str
    int
    float
    set

```python
d1 = {'a': 1, 'b': 2, 'c': 3}
d2 = {'python': 'awesome', 'matlab': 'horrible', 'a': 2}

d2.pop('a')  # remove the item associated with the key 'a'
print "Value of 'a' for d1", d1['a']  # d1 still has an 'a' item
print "Value of 'a' for d2", d2['a']  # d2 doesn't
```

###Dynamic Dispatch

>Object methods are resolved at run-time

In other words:

    some_variable.pop(0)  # what will happen?

The python interpreter calls the `.pop` method associated to the object stored in `some_variable`.  If there is no such method, an error occurs.

```python
my_dict = {0: 'a', 1: 'b', 2: 'c'}
my_list = ['a', 'b', 'c']
```

```python
# What will happen?

for obj in [my_dict, my_list]:
    print obj.pop(0)
```

###The Point:

> Object-oriented programming informs us on how we're supposed to think about code

In python, you should think in terms of:

1. Creating objects
2. Passing objects around (e.g. to functions)
3. Asking objects to do things to other objects (i.e. with their methods)
4. Asking objects to do things to themselves (again, with their methods)

In sum, your thinking should be oriented towards the **object metaphor**.


##2. Common Objects and their Uses

| I want...        | I should use...  |
| ------------- |:-------------:|
| to associate something to something else | `dict` |
| a collection of items, stored in-order | `list` |
| an ordered collection of items where item order has special significance | `tuple` |
| to store and manipulate text | `str` |
| to easily compare the contents of a collection of objects | `set` |


### A few details & caveats

```python
# Dictionaries are cool, but you need to know how they work.

my_dict = {1: 'a', 2: 'b', 3: 'c'}
for key in my_dict:  # dicts iterate over KEYS
    print key, ':', my_dict[key]
```

```python
# Tuples are immutable.  This is a good thing.

my_tuple = (True, 'incongruent')  # use tuples for heterogeneous types.
my_tuple.append(1)  # this is going to fail... hard.
```

```python
# Sets are more useful than you might think
s1 = set((1, 2, 3, 5, 6, 7, 8))
s2 = set((3, 5, 7, 8, 9, 10))

# Let's see which items are in both sets.
print s1.intersection(s2)

# Let's see which items are in one set OR the other
print s1.symmetric_difference(s2)

# Let's see all the unique items in both sets
print s1.union(s2)
```

```python
# set's can be chained!
s3 = set((1, 3, 4, 5, 9001))
print s1.symmetric_difference(s2).symmetric_difference(s3)
```
